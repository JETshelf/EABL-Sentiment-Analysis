```python
import nltk
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.tokenize import RegexpTokenizer
from  nltk import FreqDist
from nltk.stem.wordnet import WordNetLemmatizer
import re
from sklearn. feature_extraction. text import CountVectorizer,TfidfVectorizer,TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
from sklearn.pipeline import make_pipeline

import string
nltk.download('punkt')
nltk.download('stopwords')

%matplotlib inline
```

```python
df1=pd.read_csv('guinsess-tweeter.csv')
df1.head(15)
```

```python
print(df1.shape)
print(df1.info)
```

```python
df1 = df1.drop(df1.columns[0], axis=1)
df1.head
```

```python
#Checking if the dataset has any null values for every column
print("The sum of null values per column:")
print(df1.isna().sum())
print("\n")

# Checking for duplicates
print("The data has the following duplicates:")
duplicates_count = df1.duplicated().sum()
print(duplicates_count)

```

```python
#A function to remove mentions, URLs, and special characters

def clean_text(text):
    # Remove mentions, URLs, and special characters
    cleaned_text = re.sub(r'(@\S+|http\S+|[^a-zA-Z\s])', '', text)
    return cleaned_text

# Apply the cleaning function to the 'text' column
df1['cleaned_text'] = df1['text'].apply(clean_text)
```

```python
df1.head(20)
```

```python

#Applying tokenization using the nltk's word_tokenize
df1['tokenized_text'] = df1['cleaned_text'].apply(word_tokenize)
#Convert each word in the tokenized text to lowercase
df1['tokenized_text'] = df1['tokenized_text'].apply(lambda tokens: [word.lower() for word in tokens])

df1
```

```python
# Removing stop words
stop_words = set(stopwords.words('english'))
df1['filtered_text'] = df1['tokenized_text'].apply(lambda tokens: [word for word in tokens if word.lower() not in stop_words])
df1
```

```python
#Export cleaned data to a new CSV file 
columns_to_drop =df1[['stats','links','text_tokenized','tokenized_text','cleaned_text']]
df1 = df1.drop(columns=columns_to_drop)
df1
```

```python
#Export cleaned data to a new CSV file for manual labelling
df1.to_csv('cleaned_guinness_data.csv', index=False)
```
